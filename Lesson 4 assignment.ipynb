```python
# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

import pandas as pd
import numpy as np

# Your code here:
data_null = {'A': [1, 2, np.nan, 4],
             'B': [5, np.nan, 7, 8],
             'C': [9, 10, 11, np.nan]}
df_null = pd.DataFrame(data_null)
print("\nTask 3 DataFrame with null values:")
display(df_null)

df_filled = df_null.fillna(0) # Filling null values with 0
print("\nTask 3 DataFrame after filling null values:")
display(df_filled)
```

```python
# Task 4: GroupBy and Describe
# Using the following DataFrame, group by 'Category' and describe the 'Value' column

import pandas as pd

df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 15, 25, 30, 35]
})

# Your code here:
print("\nTask 4 GroupBy and Describe:")
display(df_group.groupby('Category')['Value'].describe())
```

```python
# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
df_concat = pd.concat([df1, df2], ignore_index=True)
print("\nTask 5 Concatenated DataFrame (vertically):")
display(df_concat)

df_merge = pd.merge(df_concat, df3, left_index=True, right_index=True)
print("\nTask 5 Merged DataFrame (horizontally):")
display(df_merge)
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
print("\nTask 9:")
# Iterator for the first 5 even numbers
even_numbers_list = [2, 4, 6, 8, 10]
even_iterator = iter(even_numbers_list)

print("Iterator output:")
print(next(even_iterator))
print(next(even_iterator))
print(next(even_iterator))
print(next(even_iterator))
print(next(even_iterator))

# Generator for the first 5 even numbers
def even_generator(n):
  for i in range(n):
    yield (i + 1) * 2

print("Generator output:")
even_gen = even_generator(5)
print(next(even_gen))
print(next(even_gen))
print(next(even_gen))
print(next(even_gen))
print(next(even_gen))
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Your code here:
print("\nTask 10:")
# Map to square all numbers
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared numbers using map:", squared_numbers)

# Reduce to find the product
product_of_numbers = reduce(lambda x, y: x * y, numbers)
print("Product of numbers using reduce:", product_of_numbers)

# Filter to get only even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers using filter:", even_numbers)
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

  def perimeter(self):
    return 2 * (self.length + self.width)

print("\nTask 11:")
# Create a Rectangle object
my_rectangle = Rectangle(5, 10)

# Calculate and print area and perimeter
print(f"Rectangle Area: {my_rectangle.area()}")
print(f"Rectangle Perimeter: {my_rectangle.perimeter()}")
```

```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

import pandas as pd

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
print("\nTask 12:")
# 1. Find the average salary by department
average_salary_by_department = df_employees.groupby('Department')['Salary'].mean()
print("Average salary by department:")
display(average_salary_by_department)

# 2. Get the names of employees with salary > 60000
employees_high_salary = df_employees[df_employees['Salary'] > 60000]['Name']
print("\nNames of employees with salary > 60000:")
display(employees_high_salary)

# 3. Add a new column 'Bonus' which is 10% of salary
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("\nDataFrame with 'Bonus' column:")
display(df_employees)
```
